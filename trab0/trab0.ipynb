```python
import os
from skimage import io, data_dir
from matplotlib import pyplot as plt


filename = os.path.join('./', 'baboon.png')

image = io.imread(filename)

io.imshow(image)
plt.show()
```

```python
# Adjusting the image brightness
import numpy as np

adjusted_brightness = np.round(np.power((image/255), (1/1.5)) * 255)

io.imshow(adjusted_brightness.astype(int), cmap='gray')
plt.show()
```

```python
adjusted_brightness = np.round(np.power((image/255), (1/2.5)) * 255)

io.imshow(adjusted_brightness.astype(int), cmap='gray')
plt.show()
```

```python
adjusted_brightness = np.round(np.power((image/255), (1/3.5)) * 255)

io.imshow(adjusted_brightness.astype(int), cmap='gray')
plt.show()
```

```python
# planos de bits
plano0 = np.left_shift(image, 3)
plano0 = np.right_shift(plano0, 3)

io.imshow(plano0.astype(int), cmap='gray')
plt.show()
```

```python
# planos de bits
plano1 = np.left_shift(image, 2)
plano1 = np.right_shift(plano1, 4)
plano1 = np.left_shift(plano1, 1)


io.imshow(plano1.astype(int), cmap='gray')
plt.show()
```

```python
# planos de bits
plano2 = np.left_shift(image, 1)
plano2 = np.right_shift(plano2, 6)
plano2 = np.left_shift(plano2, 2)


io.imshow(plano2.astype(int), cmap='gray')
plt.show()
```

```python
# planos de bits
plano3 = np.right_shift(image, 3)
plano3 = np.left_shift(plano3, 3)


io.imshow(plano3.astype(int), cmap='gray')
plt.show()
```

```python
# merge de imagens
filename = os.path.join('./', 'butterfly.png')

image2 = io.imread(filename)

merged = (image * 0.3) + (image2 * 0.7)

io.imshow(merged.astype(int), cmap="gray")
plt.show()
```
